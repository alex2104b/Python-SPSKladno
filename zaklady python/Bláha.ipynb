Vypište výsledek a určete, zda je číslo sudé nebo liché.


````python
#TODO
x = 3
y = 7
z = x + y
zz = z ** 2
zzz = zz ** 0.5
zzzz = zzz * 3 + 1

print("pro x=", x)
print("pro y=", y)

if(zzzz % 2) ==1:
    print(f"Výsledek je {zzzz} liché")
else:
    print(f"Výsledek je {zzzz} sudé")
````

Vytvořte program, Do kterého budete zadávat matematické operace.

Napište do programu hlášku: "Vyberte matematickou operaci: 1 pro sčítání a 2 pro odčítání". Pokud uživatel zadá 1, bude zadávat dvě čísla, která se poté sečtou. Pokud 2 tak to samé ale pro rozdíl.
Výsledek vypište ve formátu:

"první +/- druhý = výsledek"


````python
#TODO
operace = int(input("Vyberte matematickou operaci: 1 pro sčítání a 2 pro odčítání:"))
if operace == 1:
    prvni = int(input("zadejte prvni cislo: "))
    druhe = int(input("zadejte druhe cislo: "))
    vysledek = prvni + druhe    
    print(f"{prvni} + {druhe} = {vysledek} ")

elif operace == 2:
    prvni = int(input("zadejte prvni cislo: "))
    druhe = int(input("zadejte druhe cislo: "))
    vysledek = prvni - druhe       
    print(f"{prvni} - {druhe} = {vysledek} ")
````


**Refereční řešení**

```
5 + 7 = 12
```

Vytvoř program, který požádá uživatele o zadání počtu bodů (0 - 100).

Na základě počtu bodů program přiřadí známku podle těchto pravidel:

* 75 až 89 bodů: "Chvalitebně (2)"
* 60 až 74 bodů: "Dobře (3)"
* 50 až 59 bodů: "Dostatečně (4)"

Pokud uživatel zadá číslo mimo rozsah 0-100, program vypíše chybovou hlášku: "Neplatný počet bodů."

````python
#TODO
main = int(input("zadejte počet bodů (0-100"))

if main < 0 or main > 100:
    print("Neplatný počet bodů")
elif main >= 90:
    print("Známka: Výborně")
elif main >= 75:
    print ("Známka: Chvalitebně")
elif main >= 60:
    print("Známka: Dobře")
elif main >= 50:
    print("Známka: Dostatečně")
else:
    print("Známka: Nedostatečně")





````

Vytvořte aplikaci, která bude přijímat tři vstupy uživatele.

* prvním bude výška v cm
* druhý váha v kg
* třetí věk

Vypočítejte mi BMI index pro tyto údaje. BMI = (váha v kg) děleno (výška v metrech na druhou)

Výpis napište ve formátu:

Uživatel ve věku "věk" má BMI: "BMI" což znamená Nadváha/ Podváha atd.

viz hodnoty níže.
* 0 až 18,5: Hodnota podváhy
* 18,5 až 25: Hodnota ideální váhy
* 25 až 30: Hodnota nadváhy
* 30 až 40: Hodnota obezity prvního stupně
* 40 a více: Hodnota těžké obezity

````python
cm = float(input("zadejte vaší výšku:"))
kg = float(input("zadejte vaší váhu"))
vek = int(input("zadejte váš věk:"))

m = cm / 100
bmi = kg / (m ** 2)

if bmi <18.5:
    x = "Podváha"
elif bmi < 25:
    x = "Ideální"
elif bmi < 30:
    x = "Nadváha"
elif bmi < 40:
    x = "Obezita prvního stupně"
elif bmi > 40:
    x = "Těžká obezita"

print(f"uživatel ve věku {vek} má BMI: {bmi} což znamená {x}")
````


**Refereční řešení**

```
Zadejte vaší výšku v cm.: 187.7
Zadejte vaší váhu v kg.: 55.8
Zadejte vaší věk: 35

```

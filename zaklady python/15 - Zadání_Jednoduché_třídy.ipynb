Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše tomto počasí


````python
class Pocasi:
    def __init__(self, vitr: float, povetrnostni_podminky: str, teplota: float, vlhkost: float):
        self.vitr = vitr
        self.povetrnostni_podminky = povetrnostni_podminky
        self.teplota = teplota
        self.vlhkost = vlhkost

    def vypis_informace(self):
        return f'Rychlost větru: {self.vitr} km/h, Počasí: {self.povetrnostni_podminky}, Teplota: {self.teplota} °C, Vlhkost: {self.vlhkost} %'
````
